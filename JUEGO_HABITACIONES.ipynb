```python
casa_habitaciones = {}
lista_habitaciones = []
lista_personas = ["rajoy", "sanchez", "iglesias", "rivera", "casado"]
lista_llaves = []
diccionario_respuestas = {"no tengo llave": "No tengo ninguna llave"}
variable_tengo_llave = {"si", "no"}
inventario_llaves_personaje = []
import random

### TE PERMITE INTRODUCIR EL NUMERO DE HABITACIONES DEL JUEGO 

def verificador_input_entero(): 
    while True:
        try:
            cantidad_habitaciones = int(input("¿Cuántas habitaciones quieres crear? (Ingresa un número igual o mayor a 2): "))
            if cantidad_habitaciones >= 2:
                return cantidad_habitaciones
            else:
                print("Por favor, ingresa un número igual o mayor a 2.")
        except ValueError:
            print("Por favor, ingresa un número entero válido.")

cantidad_habitaciones = verificador_input_entero()
print(f"Has elegido crear {cantidad_habitaciones} habitaciones.")


situacion_inicial = {
    "habitacion_actual": "habitacion_0",
    "llaves_actuales": 0,
    "habitacion_objetivo": cantidad_habitaciones
}
situacion_actual = situacion_inicial.copy()

### DA VALOR A LAS LLAVES Y HABITACIONES Y LAS INTRODUCE EN LAS VARIABLES
for i in range(0, cantidad_habitaciones):
    numero_habitacion = "habitacion_" + str(i)
    lista_habitaciones.append(numero_habitacion)

for habitacion in lista_habitaciones:
    casa_habitaciones[habitacion] = lista_habitaciones.index(habitacion)

for i in range(0, cantidad_habitaciones):
    numero_llave = "llave_" + str(i)
    lista_llaves.append(numero_llave)

def start_game():
    print(f"Acabas de despertar en una habitación y están: {lista_personas}")
    jugar(situacion_inicial["habitacion_actual"])

def jugar(habitacion):
    situacion_actual["habitacion_actual"] = habitacion
    if situacion_actual["habitacion_actual"] == situacion_inicial["habitacion_objetivo"]:
        print("Enhorabuena llegaste a Andorra")
    else:
        print("Ahora estás en la habitación " + habitacion)
        
        while True:
            que_hago = input("¿Qué quieres hacer? Elige entre 'dialogar' o 'luchar': ").lower()
            if que_hago == "dialogar":
                if dialogar_funcion(situacion_actual["habitacion_actual"]):
                    pasar_de_habitacion()
                    pass
            elif que_hago == "luchar":
                if luchar_funcion(situacion_actual["habitacion_actual"]):
                    pasar_de_habitacion()
                    pass
            else:
                print("Opción no válida. Elige entre 'dialogar' o 'luchar'.")

def dialogar_funcion(habitacion):
    dialogar_con = input(f"¿Con quién quieres hablar? Elige entre: {', '.join(lista_personas)}: ").lower()
    if dialogar_con in lista_personas:
        tiene_llave = random.choice([True, True, False, False])
        if tiene_llave:
            print(f"{dialogar_con} tiene una llave.")
            inventario_llaves_personaje.append(lista_llaves[casa_habitaciones[habitacion]])
            print(f"Conseguiste la llave {lista_llaves[casa_habitaciones[habitacion]]}")
            return True  # Exit the loop, as the player has obtained the key
        else:
            print(f"{dialogar_con}: {diccionario_respuestas['no tengo llave']}")
            return False  # Continue the loop, as the character doesn't have the key
    else:
        print("Persona no válida.")
        return False  # Continue the loop, as the character selection is invalid

def luchar_funcion(habitacion):
    luchar_con = input(f"¿A quién quieres enfrentarte? Elige entre: {', '.join(lista_personas)}: ").lower()
    if luchar_con in lista_personas:
        print(f"Derrotaste a {luchar_con}.")
        
        # Introduce a random chance for the defeated character to have a key (25% chance)
        tiene_llave = random.choice([True, True, False, False])
        
        if tiene_llave:
            inventario_llaves_personaje.append(lista_llaves[casa_habitaciones[habitacion]])
            print(f"{luchar_con} tenía una llave. Conseguiste la llave {lista_llaves[casa_habitaciones[habitacion]]}")
            return True  # Exit the loop, as the player has obtained the key
        else:
            print(f"{luchar_con} no tenía nada.")
            return False  # Continue the loop, as the character doesn't have the key
    else:
        print("Persona no válida.")
        return False  # Continue the loop, as the character selection is invalid

def pasar_de_habitacion():
    habitacion_actual = situacion_actual["habitacion_actual"]
    if habitacion_actual == lista_habitaciones[-1]:
        print("¡HAS LLEGADO A ANDORRA!")
        exit()
    llave_actual = lista_llaves[casa_habitaciones[habitacion_actual]]
    if llave_actual in inventario_llaves_personaje:
        index_habitacion_actual = lista_habitaciones.index(habitacion_actual)
        nueva_habitacion = lista_habitaciones[index_habitacion_actual + 1]
        situacion_actual["habitacion_actual"] = nueva_habitacion
        print(f"Has pasado a la siguiente habitación: {nueva_habitacion}")
    else:
        print(f"No puedes pasar a la siguiente habitación. Necesitas la llave de {habitacion_actual}.")

# Start the game
start_game()
```
